```python
#!/usr/bin/env python3
"""
show_accuracy_and_heatmap.py

Produces and displays:
 - Accuracy graph (hit-rate / recall@k)
 - Co-occurrence heatmap (top-N items)
"""

import os
import random
from pathlib import Path
from typing import List, Dict, Any

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

# optional
try:
    import h5py
    H5PY_AVAILABLE = True
except Exception:
    H5PY_AVAILABLE = False

# ---------- CONFIG ----------
OUTPUT_DIR = Path(r"C:\Users\NXTWAVE\Downloads\Groceries Price Detection")
ARTIFACTS_PKL = OUTPUT_DIR / "artifacts.pkl"
PROCESSED_H5 = OUTPUT_DIR / "processed_data.h5"
ORIGINAL_CSV = Path(r"C:\Users\NXTWAVE\Downloads\Groceries Price Detection\archive (1)\Groceries_dataset.csv")

TOP_N_ITEMS_FOR_HEATMAP = 30
EVAL_SAMPLE_LIMIT = 3000
K_VALUES = [1, 3, 5, 10]
RANDOM_SEED = 42

random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)


# ------------------ Helpers ------------------

def load_artifacts(pkl_path: Path) -> Dict[str, Any]:
    if pkl_path.exists():
        return joblib.load(str(pkl_path))
    return {}

def load_processed_array(h5_path: Path = None) -> tuple:
    if h5_path and h5_path.exists() and H5PY_AVAILABLE:
        with h5py.File(str(h5_path), 'r') as hf:
            data = np.array(hf['data']).astype(np.uint8)
            index = [str(x) for x in hf['index'][()]]
            columns = [str(x) for x in hf['columns'][()]]
        return data, index, columns
    raise FileNotFoundError("No processed_data.h5 found.")

def rebuild_basket_from_csv(csv_path: Path) -> tuple:
    df = pd.read_csv(str(csv_path), encoding='latin-1', low_memory=False)
    grouped = df.groupby(['Member_number', 'Date'])['itemDescription'].apply(lambda s: list(s.astype(str).str.strip().unique()))
    transactions = grouped.tolist()
    index = grouped.index.tolist()
    from sklearn.preprocessing import MultiLabelBinarizer
    mlb = MultiLabelBinarizer(sparse_output=False)
    data = mlb.fit_transform(transactions).astype(np.uint8)
    columns = list(mlb.classes_)
    index = [str(x) for x in index]
    return data, index, columns

def make_recommender_from_rules_list(rules_list: List[Dict[str, Any]]):
    rules_internal = []
    for r in rules_list:
        rules_internal.append({
            'antecedents': set(r.get('antecedents', [])),
            'consequents': list(r.get('consequents', [])),
            'lift': r.get('lift', 0.0)
        })
    def recommend(given_items, top_n=5):
        if isinstance(given_items, str):
            given = {given_items}
        else:
            given = set(given_items)
        scores = {}
        for r in rules_internal:
            if r['antecedents'].issubset(given):
                for c in r['consequents']:
                    scores[c] = max(scores.get(c, 0.0), r.get('lift', 0.0))
        for g in list(given):
            scores.pop(g, None)
        sorted_items = sorted(scores.items(), key=lambda x: x[1], reverse=True)
        return [item for item, sc in sorted_items[:top_n]]
    return recommend

def prepare_transactions_list(data_np: np.ndarray, columns: list) -> list:
    tx_list = []
    for i in range(data_np.shape[0]):
        idx = np.where(data_np[i] > 0)[0]
        items = [columns[j] for j in idx]
        tx_list.append(items)
    return tx_list

def evaluate_hit_rate(rules_serial, transactions_list, ks, sample_limit=None):
    recommender = make_recommender_from_rules_list(rules_serial)
    valid_idx = [i for i, t in enumerate(transactions_list) if len(t) >= 2]
    if sample_limit and len(valid_idx) > sample_limit:
        valid_idx = list(np.random.choice(valid_idx, size=sample_limit, replace=False))
    hits = {k: 0 for k in ks}
    total = 0
    for idx in valid_idx:
        t = transactions_list[idx]
        hold = random.choice(t)
        given = [x for x in t if x != hold]
        if not given:
            continue
        total += 1
        for k in ks:
            recs = recommender(given, top_n=k)
            if hold in recs:
                hits[k] += 1
    return {k: hits[k]/total if total>0 else 0.0 for k in ks}

def compute_cooccurrence_matrix(data_np, columns, top_n=30):
    counts = data_np.sum(axis=0)
    top_idx = np.argsort(-counts)[:top_n]
    top_cols = [columns[i] for i in top_idx]
    sub = data_np[:, top_idx].astype(np.uint8)
    cooc = np.dot(sub.T, sub)
    return pd.DataFrame(cooc, index=top_cols, columns=top_cols)

def plot_accuracy(hit_rates: dict):
    ks = sorted(hit_rates.keys())
    vals = [hit_rates[k] for k in ks]
    plt.figure(figsize=(6,4))
    plt.plot(ks, vals, marker='o')
    plt.xticks(ks)
    plt.xlabel("k (top-k)")
    plt.ylabel("Hit-rate (Recall@k)")
    plt.title("Recommender Hit-rate vs k")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

def plot_heatmap(df_cooc):
    plt.figure(figsize=(12,10))
    sns.heatmap(df_cooc, cmap="YlGnBu", annot=False)
    plt.title("Item Co-occurrence Heatmap (top items)")
    plt.show()


# ------------------ Main ------------------

def main():
    # 1) Load artifacts
    artifacts = load_artifacts(ARTIFACTS_PKL)
    rules_serial = artifacts.get('rules_serial', None)
    
    # 2) Load processed data or rebuild
    try:
        data_np, index, columns = load_processed_array(PROCESSED_H5)
        print(f"Loaded processed_data (tx={data_np.shape[0]}, items={data_np.shape[1]})")
    except Exception as e:
        print(f"Processed data not found. Rebuilding from CSV: {e}")
        data_np, index, columns = rebuild_basket_from_csv(ORIGINAL_CSV)
        print(f"Rebuilt basket (tx={data_np.shape[0]}, items={data_np.shape[1]})")
    
    # 3) Load rules from JSON if artifacts missing
    if rules_serial is None:
        rules_json_path = OUTPUT_DIR / "association_rules.json"
        if rules_json_path.exists():
            import json
            with open(str(rules_json_path), 'r') as fh:
                rules_serial = json.load(fh)
            print("Loaded rules from JSON")
        else:
            print("No rules found. Exiting.")
            return
    
    # 4) Prepare transactions
    transactions_list = prepare_transactions_list(data_np, columns)
    
    # 5) Evaluate hit-rate
    hit_rates = evaluate_hit_rate(rules_serial, transactions_list, ks=K_VALUES, sample_limit=EVAL_SAMPLE_LIMIT)
    print("Hit-rate@k:", hit_rates)
    
    # 6) Plot accuracy
    plot_accuracy(hit_rates)
    
    # 7) Compute & plot co-occurrence heatmap
    df_cooc = compute_cooccurrence_matrix(data_np, columns, top_n=TOP_N_ITEMS_FOR_HEATMAP)
    plot_heatmap(df_cooc)

if __name__ == "__main__":
    main()
```
